```python
import pandas as pd
import numpy as np
```

```python
attendance = pd.read_csv("church_attendance.csv")
attendance
```

```python
#clean column names
columns = attendance.columns.to_list()
columns [0] = 'First_Name'
columns [1] = 'Last_Name'
columns [3] = 'House_Hold_Number'
columns [4] = 'Service_attended'
attendance.columns = columns
attendance
```

```python
#Extracting columns
attendance['First_Name']
```

```python
attendance[['First_Name', 'House_Hold_Number']]
```

```python
gender = ["Male", "Female", "Male", "Male", "Male"]
```

```python
attendance
```

```python
attendance['Gender'] = gender
attendance
```

```python
#Edit a column
#Add 4 to service attended
attendance['Service_attended'] = attendance['Service_attended'] + 4
attendance
```

```python
#delete a column

attendance.pop("Gender")
```

```python
attendance
```

```python
#create a new dataframe of worker status and age
worker_status = attendance['Worker_Status']
age = [18,20,22,24,26]
new_df = pd.DataFrame({"worker_status": worker_status,  "Age": age})
new_df
```
